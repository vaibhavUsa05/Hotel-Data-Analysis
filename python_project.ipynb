```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="adr",width=141.2)
plt.title("Boxplot of ADR")
plt.xlabel("ADR")
plt.show()
```

```python
q1=df["adr"].quantile(0.25)
print(q1)
```

```python
q3=df["adr"].quantile(0.75)
print(q3)
```

```python
IQR=q3-q1
lower_bound=q1-1.5*q3
upper_bound=q1+1.5*q3
print(lower_bound,upper_bound)
```

```python
# taking the 25% to 75%-ile values and replacig the outlier values with the median of the df["adr"] values
df.loc[(df["adr"]<lower_bound) | (df["adr"]>upper_bound), "adr"]= df["adr"].median()
```

```python

sns.boxplot(data=df,x="adr")
plt.xlabel("The modified values of adr")
plt.ylabel("outlier")
plt.title("The modified graph od adr after removing outliers")
plt.show()
```

```python
total_canceled_reservations= df["is_canceled"].value_counts(normalize=True)
total_canceled_reservations
```

```python
plt.figure(figsize=(8, 6))
graph_legend = sns.countplot(x="hotel", hue="is_canceled", color="blue", data=df, width=0.7)
legend_labels = graph_legend.get_legend_handles_labels()
plt.legend(legend_labels[0], ["Canceled", "Not Canceled"])
plt.xlabel("Hotels")
plt.ylabel("Total number of reservations")
plt.title("Reservation status in different hotels")
plt.show()
```

```python
resort_hotels=df[df["hotel"]=='Resort Hotel']
hotels_percentage=resort_hotels["is_canceled"].value_counts(normalize=True)
hotels_percentage
```

```python
city_hotels=df[df["hotel"]=='City Hotel']
hotels_percentage=city_hotels["is_canceled"].value_counts(normalize=True)
hotels_percentage
```

```python
resort_hotels_mean = resort_hotels["adr"].mean()
city_hotels_mean = city_hotels["adr"].mean()
print( city_hotels_mean)
```
